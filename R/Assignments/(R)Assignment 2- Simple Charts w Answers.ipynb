## 1.) Pie charts
Make a pie chart of the percentage of time you spend during the day on various activities (sleeping, eating, socializing, attending class, homework, etc).

```r
lbls <- c("Sleeping", "Eating", "Working Out", "School", "TV", "Free Time")
slices <- c(33, 10, 13, 21, 10, 13)
pct <- round(slices/sum(slices)*100)

lbls <- paste(lbls, pct,sep=" ")
lbls <- paste(lbls, "%")

pie(slices, labels = lbls, col=rainbow(length(lbls)),
   main="How I Spend My Day")
```

## 2.) Bar Plots

Make a bar plot of the total amount of time you spend during the day on various activities (sleeping, eating, socializing, attending class, homework, etc). Be sure to include vertical or slanted labels for each bar on the X axis.

```r
lbls<-c("Sleeping", "Eating", "Working Out", "School", "TV", "Free Time")
time<-c(8, 2, 4, 4, 2, 4)
barplot(time,names.arg = lbls, xlab="Activity", ylab="Hours",
       main="How I Spend my Day", col=rainbow(length(lbls)), border = "red")
```

## 3.) Group bar plots with legend
Make either a side-by-side or stacked bar plot with the total amount of time you spend during the day on weekdays versus weekends. 

```r
install.packages("ggplot2")
library(ggplot2)

weekend<-read.table(header=TRUE, text='
        Time      Activity     Hours
    1   Week      Sleeping     8
    2   Weekend   Sleeping     8
    3   Week      Eating       2
    4   Weekend   Eating       3
    5   Week      WorkingOut   4
    6   Weekend   WorkingOut   2
    7   Week      School       4
    8   Weekend   School       0
    9   Week      TV           2
    10  Weekend   TV           3
    11  Week      FreeTime     4
    12  Weekend   FreeTime     8
') 
ggplot(weekend, aes(Activity, Hours, fill = Time)) + 
  	geom_bar(stat="identity", position = "dodge") + 
  	scale_fill_brewer(palette = "Set3") + ggtitle("How I Spend my Time")
```
