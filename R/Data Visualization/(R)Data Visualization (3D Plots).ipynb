## 1. Contour Plot

We will also be using a new dimension in our 3D plots. You may be used to just using x and y as coordinates, but for a 3D plot, it is necessary to incorporate z  coordinates. 
X and Y are called “predictor variables”, while Z is referred to as the “response variable”. Predictor variables are very similar to independent variables, as response variables are very similar to dependent variables.

Now, let’s begin our contour plot. A contour plot is a graphical technique for representing a 3D surface by plotting z slices, called “contours”, on a 2D format. That is, given a value for z, lines are drawn for connecting the (x,y) coordinates where that z value occurs.

We can create a contour plot using [contour()]( https://www.rdocumentation.org/packages/graphics/versions/3.5.0/topics/contour), a generic function in R.

```r
x <- seq(-1, 1, len=100)
y <- seq( -2, 2, len=100)
z <- outer(x, y, function(x, y) x^2+y^2) 
contour(x, y, z, nlevels=12, main = "A simple example: z = x^2+y^2", col = "red", xlab = "x", ylab = "y")
```

We can also use [ggplot2]( https://ggplot2.tidyverse.org/reference/geom_contour.html) to make our contour plot. You need to install the package if you haven't done it before:

```r
install.packages("ggplot2")
```

The data needs to be in the form of a data.frame. We will learn more about dataframes in the next section.

```r
library(ggplot2)
# Create some data sets we can graph...
twoDimData  <-expand.grid(x=seq(from=-3, to=3, length.out=30), y=seq(from=-3, to=3, length.out=30))
twoDimData$z<-sin(twoDimData$x*twoDimData$x+twoDimData$y*twoDimData$y)
# Plot data
ggplot(data=twoDimData, aes(x=x, y=y)) +
    geom_tile(aes(fill=z))
```

## 2.Surface Plot
Our goal for this portion is to create a surface plot that has a bump in the middle. You might be wondering, what is a surface plot? Essentially, it is a diagram of three-dimensional data. Rather than showing individual data points, surface plots show a functional relationship between a dependent variable (Z), and two independent variables (X and Y).
 
There are many packages in R for 3D plotting. The following example is using latticeExtra::wireframe.

```r
install.packages("latticeExtra")
```

```r
library(latticeExtra)
z <- as.matrix(read.table(text='-4  -3  -2  -1   0   1   2  3   4
1 159 144 133 132 138 123  80 28 -22
2 153 135 121 122 160 162 110 50  -7
3 148 126 107 104 161 190 135 67   3
4 145 120  96  92 161 202 149 77   8
5 144 117  92  89 161 205 153 80  10
6 145 120  96  92 161 202 149 77   8
7 148 126 107 104 161 190 135 67   3
8 153 135 121 122 160 162 110 50  -7
9 159 144 133 132 138 123  80 28 -22', header=TRUE, check.names=FALSE))

wireframe(z, scales=list(arrows=FALSE), xlab='x', ylab='y', main="An Example Surface Plot", drape=TRUE,
          col.regions=terrain.colors(10), at=seq(min(z), max(z), len=11))
```
