## Pie Chart

This link describes how to create a pie chart in R, https://www.statmethods.net/graphs/pie.html. 

It shows how to create a simple pie chart and a pie chart with annotated percentages, which is the technique we used in our example. However, feel free to experiment with the different ways to gain more experience!

```r
# Let's create a pie chart.

# Plot and label data.
slices <- c(60.0, 12.9, 5.7, 16.3, 5.0) 
lbls <- c("White", "Asian", "African American", "Hispanic", "Other")
pct <- round(slices/sum(slices)*100)

# Add percentages to the labels.
lbls <- paste(lbls, pct)

# Add the percent sign to the labels.
lbls <- paste(lbls,"%",sep="") # ad % to labels 

# Make labels, titles, and a color scheme.
pie(slices,labels = lbls, col=rainbow(length(lbls)),
    main="Loyola's Diversity: Undergrad Students (Fall 2016)")
```

<center>Figure R.1. : Loyola's racial diversity represented by a pie chart</center>

## Bar Plot

```r
# Let's create a bar graph.

# Type in data points.
Price <- c(51273 , 58602)
Year <- c("2013-2014","2016-2017")

# Plot the chart.
barplot(Price,names.arg = Year, xlab = "Year", ylab = "Estimated Cost ($)", col = "mistyrose",
        main = "Costs of Living On Campus at Loyola", border = "purple")

# Add labels.
text(0.71, 48000, "$51,273")
text(1.9, 55000, "$58,602")
```

<center> Figure R.2: Cost of on campus-living per year</center>

## Side-by-side Bar Plot

Here is the link we used to create our group bar chart example: https://stackoverflow.com/questions/17721126/simplest-way-to-do-grouped-barplot. 

In the first answer to the question, it conveys the use of the ggplot function. In the other answers, it also shows other ways to create group bar charts with R, so you can try other those out if you would like!

```r
# Let's create a group bar chart. 

# Import the necessary package.   
install.packages("ggplot2")
library(ggplot2)

Immigration <- read.table(
  header=TRUE, text='Legend        Year Percent
1  Other         1980       3
  2  Mexican       1980      16
  3  Other         1990       6
  4  Mexican       1990      22
  5  Other         2000       7
  6  Mexican       2000      30
  7  Other         2010       8
  8  Mexican       2010      29')

ggplot(Immigration, aes(factor(Year), Percent, fill = Legend)) + 
  geom_bar(stat="identity", position = "dodge") + 
  scale_fill_brewer(palette = "Set3") + ggtitle("Demographics of Latin American US Immigrants")
```

This link shows how to create multiple bar plots in R, including a stacked bar plot: https://www.statmethods.net/graphs/bar.html. 

```r
# Let's create a stacked bar graph.

# Enter labels and data points. 
dat <- read.table(text = "South-Lawndale    Englewood    Austin    Albany-Park    Bridgeport    Rogers-Park    Lincoln-Park  
HardshipIndex 96 94 73 53 43 39 02
PercentAgedwithoutHighSchoolDiploma 55 29 24 33 22 18 04", header = TRUE)

# Set up axis and legend.
barplot(as.matrix(dat), ylab="Count", xlab="Neighborhood", main="Chicago Data Portal's Annual Assesment on Neighborhoods", legend=c("Hardship Index", "% of People 25+ without High School Diploma"))
```
