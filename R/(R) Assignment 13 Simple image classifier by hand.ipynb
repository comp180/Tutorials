## 2. Read in and convert forest1.jpg into an array and print the dimensions of the array (e.g. width x height x 3)

Recall that (raster-based) images are defined by an array of pixels with dimensions and width and height.  Grayscale images only need one number per pixel (luminance), but color images need three numbers per pixel. Often those color components are RGB (red, green, blue) but other designations are possible (If you see 4 layers, see **Transparency** in troubleshooting).

Interesting note on RGB color values. Unlike when you mixed paints as a kid, color values mix light (additive colors rather than subtractive). Long story short, if you want to have an intuition of color in images its important to note how colors mix below.

Interesting mixes to note in this assignment: 
    - yellow colors (like sand) are actually composed of red and green.
    - sky blue (e.g. cyan) may have a mix of blue and green.
    - Brown colors are a slight mix of mostly red, followed by green, and a bit of blue.
    - Bright colors and white are actually composed of all three colors
    - Black or dark colors have very small values for all colors



**Additive vs Subtractive colors:** RGB values in defining pictures are additive colors, which behave differently in combination than mixing the same colors with paints (subtractive colors).

## Troubleshooting:

**Transparency:** Some images will by width x height x 4 instead of x 3. This is because some image types include a fourth number to define the amount of transparency in the image - referred to as the alpha. Pixels with alpha = 0 are opaque, while those with value 1 are effectively invisible.
