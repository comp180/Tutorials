## 1. Pie Chart

We are going to make a pie chart that portrays the “Percentage of Internet Users on Various Social Media Apps”. This code will define the colors and sizes of the various “slices” of the pie chart. It will also label the values on the slices and show the figure. 

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
```

[This link](https://medium.com/python-pandemonium/data-visualization-in-python-pie-charts-in-matplotlib-c71fb0fe5c56) will help describe the parameters that are used in the [plt.pie(....)](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.pie.html) function. 

In this example, we used our own data for plotting the pie chart, but you can use whichever data you want to visualize. Feel free to change the colors too, and you can even change the number of decimals that are used in the percentages by altering the "autopct" part of the pie function.  

```python
#Data to plot
labels = ['Facebook','Instagram','Twitter','LinkedIn']
sizes = [48.17,19.5,14.63,17.68]
colors = ['blue', 'purple', 'lightsteelblue','turquoise']
explode = [0.1,0,0,0] #explode 1st slice

#plotting the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%',shadow=True,startangle=270)
plt.title('Percentage of Internet Users on Various Social Media Apps',fontweight='bold')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

## 2. Bar Plot

We are going to make a bar plot that portrays the “Daily Social Media Use per Application”. This code will plot the defined data and create x- and y-labels for the graph. It will also create x- and y-ticks and make sure they line up below the various bars. 

* This link will help clarify the [bar(...)](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.bar.html) function that is used to create the bar plot, . We used align = 'center', because it will center the base on the x-positions. 

* For the [np.arange(...)](https://docs.scipy.org/doc/numpy-1.14.0/reference/generated/numpy.arange.html) function, this link explains how the function will create evenly-spaced locations for the values on the x-axis. Therefore, we use the number of values on the x-axis to determine the location of the labels on the x-axis. 

* For the [xticks(...)](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.xticks.html) function, this link helps describe how the function sets the locations and labels of the values on the x-axis.

This link explains how to [rotate the text](https://stackoverflow.com/questions/10998621/rotate-axis-text-in-python-matplotlib/37708190) on either the y-axis or x-axis by including the 'rotation' variable as a parameter of the <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.xticks.html"> xticks()</a> or <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.yticks.html"> yticks() </a> functions.  

```python
#Set x-axis labels and data
objects = ('Facebook', 'Instagram', 'Twitter', 'Snapchat','YouTube')
y_pos = np.arange(len(objects))
performance = [35,15,1,25,40]

#Create bar plot with labels, a title, and color
plt.bar(y_pos, performance, align='center', alpha=0.5,color='yellow')
#Places the labels on the x-axis at their various locations determined by np.arange(len(objects))
plt.xticks(y_pos, objects, rotation='70')
plt.ylabel('Daily Social Media Use (minutes)')
plt.xlabel('Applications')
plt.title('Daily Social Media Use Per Application')

#Display the bar plot
plt.show()
```

## 3. Grouped Bar Plot

We are going to make a side-by-side bar plot that portrays the “Number of Sales in July and August per Week”. This code will define the data for the various bars and create a legend to show which color bar corresponds to which set of data. It will also create x- and y- labels for the plot as well as a title!

* This link shows the example that was used to create this [side-by-side bar plot](https://matplotlib.org/examples/api/barchart_demo.html). In this example, we created the first set of data and then plotted the bars for that set of data. We then repeated these steps for the second set of data. Then, we set the locations of the labels on the x-axis for both of the bar plots using the x-ticks function, since both bar plots have to use the same x-axis values. 

* We also used the [legend function](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html), which took as its parameters: the numerical data from both sets of bar plots and the labels to be used in the legend for the two bar plots.


```python
#First set of data
july_sales = (500, 800, 700, 900, 750)
ind = np.arange(5)  # the x locations for the groups
width = 0.35       # the width of the bars
fig, ax = plt.subplots()

#Creating the first set of bars for the first set of data
rects1 = ax.bar(ind, july_sales, width, color='blue') 

#Second set of data-- creation of bars 
august_sales = (800, 900, 600, 800, 1000)
rects2 = ax.bar(ind + width, august_sales, width, color='orange')

# Add some text for labels, title and axes ticks
ax.set_ylabel('Number of Sales',fontsize=13)
ax.set_title('Number of Sales in July and August per Week',fontsize=13)
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('Week 1', 'Week 2', 'Week 3', 'Week 4', 'Week 5'),fontsize=10)

#Create a legend 
ax.legend((rects1[0], rects2[0]), ('July', 'August'))

plt.show()
```

<b> For the second side-by-side bar plot example, we created the autolabel(rects) function to label the various bars with their numerical data points, so it can be clarified where the bars reach on the y-axis. </b>

```python
#First set of data
july_sales = (500, 800, 700, 900, 750)
ind = np.arange(5)  # the x locations for the groups
width = 0.35       # the width of the bars
fig, ax = plt.subplots()

#Creating the first set of bars for the first set of data
rects1 = ax.bar(ind, july_sales, width, color='blue') 

#Second set of data-- creation of bars 
august_sales = (800, 900, 600, 800, 1000)
rects2 = ax.bar(ind + width, august_sales, width, color='orange')

# Add some text for labels, title and axes ticks
ax.set_ylabel('Number of Sales',fontsize=13)
ax.set_title('Number of Sales in July and August per Week',fontsize=13)
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('Week 1', 'Week 2', 'Week 3', 'Week 4', 'Week 5'),fontsize=10)

#Create a legend 
ax.legend((rects1[0], rects2[0]), ('July', 'August'))

#Attach a text label above each bar displaying its height
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., height,
                '%d' % int(height),
                ha='center', va='bottom')
autolabel(rects1)
autolabel(rects2)
plt.show()
```

## 3a. Stacked Bar Plot


We are going to make a stacked bar plot that portrays the “Number of Sales in July and August per Week”. This code will define the data sets to plot and create the bars that will be stacked on top of each other. It will also create x- and y- labels, a title, and tick marks. This code creates labels for each bar as well as a legend, so the number of sales per week can be shown clearly. 

Here is the link that we used as an aid to [create our stacked bar plot](https://matplotlib.org/gallery/lines_bars_and_markers/bar_stacked.html). Similar to creating the side-by-side bar plot, we created a set of data for each bar plot and then used the bar() function to plot the bars for the data. 

<b>For the second stacked bar plot example, we used a similar method that's used in the autolabel() function in the side-by-side bar plot example to label each bar with its numerical data point to clarify the location of the bar on the y-axis. </b>

```python
#Data for the first and second bar plots
july_sales = (500, 800, 700, 900, 750)
august_sales = (800, 900, 600, 800, 1000)

#Setting the width of the bars and tick positions
width = 0.5      # the width of the bars
ind = np.arange(1,6)  # the x locations for the groups
tick_pos = [i + (width/50) for i in ind]

#Creating the bars
p1 = plt.bar(ind, july_sales, width, color='blue',align='center')
p2 = plt.bar(ind,august_sales,width,bottom=july_sales,color='green',align='center')

#Add some text for labels, title, axes ticks, and legend
plt.ylabel('Number of Sales',fontsize=12)
plt.xlabel('Weeks',fontsize=12,labelpad=15)
plt.title('Number of Sales in July and August per Week',fontsize=13)
plt.xticks(tick_pos,('Week 1','Week 2','Week 3','Week 4','Week 5'), fontsize=10)
plt.legend((p1[0], p2[0]), ('July', 'August'),loc="best")

#Display the bar plot
plt.show()
```

```python
#Data for the first and second bar plots
july_sales = (500, 800, 700, 900, 750)
august_sales = (800, 900, 600, 800, 1000)

#Setting the width of the bars and tick positions
width = 0.5      # the width of the bars
ind = np.arange(1,6)  # the x locations for the groups
tick_pos = [i + (width/50) for i in ind]

#Creating the bars
p1 = plt.bar(ind, july_sales, width, color='blue',align='center')
p2 = plt.bar(ind,august_sales,width,bottom=july_sales,color='green',align='center')

#Add some text for labels, title, axes ticks, and legend
plt.ylabel('Number of Sales',fontsize=12)
plt.xlabel('Weeks',fontsize=12,labelpad=15)
plt.title('Number of Sales in July and August per Week',fontsize=13)
plt.xticks(tick_pos,('Week 1','Week 2','Week 3','Week 4','Week 5'), fontsize=10)
plt.legend((p1[0], p2[0]), ('July', 'August'),loc="best")

#Labels for each bar
for r1, r2 in zip(p1, p2):
    h1 = r1.get_height()
    h2 = r2.get_height()
    plt.text(r1.get_x() + r1.get_width() / 2., h1 / 2., "%d" % h1, ha="center", va="bottom", color="white", fontsize=10, fontweight="bold")
    plt.text(r2.get_x() + r2.get_width() / 2., h1 + h2 / 2., "%d" % h2, ha="center", va="bottom", color="white", fontsize=10, fontweight="bold")

#Display the bar plot
plt.show()
```

## 4. Histogram

This link was used to create both examples for making a [histogram with Python](https://pythonspot.com/matplotlib-histogram/). 

<b> The first example creates a simple histogram without a best fit line. The <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html"> hist() </a>function takes in the values that are going to be sorted into bins and plotted in the histogram. The function also lets you choose a color for the bars, an alpha value and an edge color. </b>

To generate random data that is normally distributed, we want to use numpy's random class, which contains a <a href="https://docs.scipy.org/doc/numpy-1.14.0/reference/generated/numpy.random.normal.html"> normal()</a> function to "draw random samples from a normal distribution". 

```python
x = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]
num_bins =5
n, bins, patches = plt.hist(x, num_bins, facecolor='blue', alpha=0.75, edgecolor='black')

plt.title('Simple Histogram')
plt.xlabel('X-values',fontsize=14)
plt.ylabel('Y-values',fontsize=14)

plt.show()
```

<b> In this second example, we made a histogram that portrays various IQs and the probability of people having these IQs. This code will create a best-fit line for the histogram and add it to the plot as soon as the data has been created. A grid and labels will also be added to the plot.  </b>

```python
#Used to create a best-fit line
mu, sigma = 100, 15
x = mu + sigma*np.random.randn(10000)

#Creates a histogram of the data
#This histogram has 30 bins, but it can be changed.
n, bins, patches = plt.hist(x, 30, normed=1,facecolor='green', alpha=0.75,edgecolor='black')

#Adds a 'best fit' line
y = mlab.normpdf(bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=1)

#Adds x- & y-labels for the history and creates a grid
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=15$')
#Highlighted code below
plt.axis([40, 160, 0, 0.03]) #Gives ranges for x & y values
#End of highlighted code
plt.grid(True)

#Shows the histogram
plt.show()
```
