## Multiple Line Plot

Letâ€™s make a multiple line plot! The code below creates multiple sets of data from which lines will be created for the plot. This code also creates labels and a legend for the plot. 

This link describes how to create a plot with multiple lines, and it provides examples of customizing the plot with different line styles, color styles, and more: https://www.kaggle.com/andyxie/matplotlib-plot-multiple-lines/code. 

The following link shows the different marker styles that can be used in the plot function: https://matplotlib.org/api/markers_api.html. 

```python
import matplotlib.pyplot as plt
#Varying y-values and x-values
july_sales = [500,800,700,900,750,600,400]
august_sales = [800,900,600,800,1000,700,650]
days_of_the_week = [1,2,3,4,5,6,7]

#Setting x-labels
week_ticks = ['Week 1','Week 2','Week 3','Week 4','Week 5','Week 6','Week 7']

#Plotting the two different lines
plt.plot(days_of_the_week,july_sales,marker='o',color='b')
plt.plot(days_of_the_week,august_sales,marker='v',color='r')

#Putting a title and labels
plt.title("July Sales vs. August Sales Per Week")
plt.ylabel("Number of Sales")
plt.legend(['July Sales','August Sales'],loc = 'lower center')
plt.xticks(days_of_the_week,week_ticks)
plt.show()
```

## Scatter Plot

Letâ€™s make a scatter plot! The code below creates multiple sets of data from which data points will be set on the figure.

The second example creates a function to calculate a line of best-fit and adds the line to the scatter plot. 

The following link explains the scatter function and its parameters: /matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html. 

The following link also explains how to create a scatter plot with either the plot or scatter function: https://jakevdp.github.io/PythonDataScienceHandbook/04.02-simple-scatter-plots.html. Feel free to try both ways if you'd like to learn the different techniques of creating a scatter plot. 

```python
import numpy as np
import matplotlib.pyplot as plt
# Fixing random state for reproducibility
np.random.seed(19680801)

#Getting x and y values
temperatures = [68,63,74,72,79,78,71,71,69,66,80,58]
customers = [317,355,463,419,507,431,433,434,362,340,540,250]

#Creating the scatter plot
plt.scatter(temperatures, customers, s=100, facecolors ='none', edgecolors='black', alpha=0.5)

#Labelling the plot
plt.xlabel('Average Temperature at the Pool (Degrees Fahrenheit)')
plt.ylabel('Number of Customers')
plt.title('Number of Customers vs. Average Temperature at the Pool')

#Outputting the scatter plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Fixing random state for reproducibility
np.random.seed(19680801)

#Getting x and y values
temperatures = [68,63,74,72,79,78,71,71,69,66,80,58]
customers = [317,355,463,419,507,431,433,434,362,340,540,250]

#Function for line of best fit
def best_fit(X, Y):
    xbar = sum(X)/len(X)
    ybar = sum(Y)/len(Y)
    n = len(X) # or len(Y)
    numer = sum([xi*yi for xi,yi in zip(X, Y)]) - n * xbar * ybar
    denum = sum([xi**2 for xi in X]) - n * xbar**2
    b = numer / denum
    a = ybar - b * xbar
    print('best fit line:\ny = {:.2f} + {:.2f}x'.format(a, b))
    return a, b

#Calculating the line of best fit
a, b = best_fit(temperatures, customers)

#Creating the scatter plot
plt.scatter(temperatures, customers, s=100, facecolors ='none', edgecolors='black', alpha=0.5)
yfit = [a + b *xi for xi in temperatures]

#Labelling the plot
plt.xlabel('Average Temperature at the Pool (Degrees Fahrenheit)')
plt.ylabel('Number of Customers')
plt.title('Number of Customers vs. Average Temperature at the Pool')
plt.plot(temperatures,yfit)

#Outputting the scatter plot
plt.show()
```

## Subplot <br>

<font color = "red">Note, the syntax used for the subplot command below is not the most common. To explain, fig.add_subplot(221) divides the figure into two rows (the first 2) and two columns (the second 2) and selected the first of 4 possible locations. fig.add_subplot(2,2,1) works similarly. However, you also have to specify the axes for each separate plot.
More typically you see people just prepare an axis using
plt.subplot(2, 1, 1)
and then simply plotting as they normally would after each subplot command.
More examples: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html </font>

Letâ€™s make a subplot! The code below creates multiple sets of data from which multiple line plots will be made. It also creates different axes for the line plots and puts them into the same figure. It also creates labels for each of the plots. 

The first example creates the subplots as line plots for when the data points are ordered; the second example creates the subplots as scatter plots, since the data points are randomly ordered. 

```python
import random
import matplotlib.pyplot as plt
from matplotlib import style

#Setting the style of the subplot
style.use('fivethirtyeight')
fig = plt.figure()

#Creating data
#First plot:
average_temp_in_june = [70,70,74,80]
customers_in_july = [237,250,325,457]

#Second plot:
average_temp_in_july = [76,78,80,82]
customers_in_august = [480,489,498,500]

#Third plot
average_temp_in_august = [71,73,80,86]
customers_in_august = [410,500,600,615]

#Returns the instance of the Axes
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(212)

#Creates one plot with random x & y values
ax1.set_ylabel('Amount of customers',fontsize=8)
ax1.set_xlabel('Average Temps in June (Degrees Fahrenheit)',fontsize=8)
ax1.plot(average_temp_in_june,customers_in_july)

#Creates a second plot with random x & y values
ax2.set_ylabel('Amount of customers',fontsize=8)
ax2.set_xlabel('Average Temps in July (Degrees Fahrenheit)',fontsize=8)
ax2.plot(average_temp_in_july,customers_in_july)

#Creates a third plot with random x & y values
ax3.set_ylabel('Amount of customers',fontsize=8)
ax3.set_xlabel('Average Temps in August (Degrees Fahrenheit)',fontsize=8)
ax3.plot(average_temp_in_august,customers_in_august)

#Labelling the subplots and creating space 
fig.suptitle('Average Temperatures vs. Amount of Customers at the Pool',fontsize=12)
fig.subplots_adjust(hspace=0.2,top=0.5,wspace=0.2)
fig.tight_layout()


#Puts all of the plots in the same figure 
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
from matplotlib import style

#Setting the style of the subplot
style.use('fivethirtyeight')
fig = plt.figure()

#Creating data
#First plot:
average_temp_in_june = [70,74,80,70]
customers_in_july = [250,325,457,237]

#Second plot:
average_temp_in_july = [82,76,78,80]
customers_in_august = [500,480,489,498]

#Third plot
average_temp_in_august = [80,71,86,73]
customers_in_august = [600,410,615,500]

#Returns the instance of the Axes
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(212)

#Creates one plot with random x & y values
ax1.set_ylabel('Amount of customers',fontsize=8)
ax1.set_xlabel('Average Temps in June (Degrees Fahrenheit)',fontsize=8)
ax1.scatter(average_temp_in_june,customers_in_july)

#Creates a second plot with random x & y values
ax2.set_ylabel('Amount of customers',fontsize=8)
ax2.set_xlabel('Average Temps in July (Degrees Fahrenheit)',fontsize=8)
ax2.scatter(average_temp_in_july,customers_in_july)

#Creates a third plot with random x & y values
ax3.set_ylabel('Amount of customers',fontsize=8)
ax3.set_xlabel('Average Temps in August (Degrees Fahrenheit)',fontsize=8)
ax3.scatter(average_temp_in_august,customers_in_august)

#Labelling the subplots and creating space 
fig.suptitle('Average Temperatures vs. Amount of Customers at the Pool',fontsize=12)
fig.subplots_adjust(hspace=1.0,top=1.0,wspace=1.0)
fig.tight_layout()

#Puts all of the plots in the same figure 
plt.show()
```
