## Pie Chart

We are going to make a pie chart that portrays the “Percentage of Internet Users on Various Social Media Apps”. This code will define the colors and sizes of the various “slices” of the pie chart. It will also label the values on the slices and show the figure. 

```python
import matplotlib.pyplot as plt

#Data to plot
labels = ['Facebook','Instagram','Twitter','LinkedIn']
sizes = [48.17,19.5,14.63,17.68]
colors = ['blue', 'purple', 'lightsteelblue','turquoise']
explode =(0.1,0,0,0) #explode 1st slice

#plotting the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%',shadow=True,startangle=270)
plt.title('Percentage of Internet Users on Various Social Media Apps',fontweight='bold')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

## Bar plot

We are going to make a bar plot that portrays the “Daily Social Media Use Per Application”. This code will plot the defined data and create x- and y-labels for the graph. It will also create x- and y-ticks and make sure they line up below the various bars. 

```python
import numpy as np
import matplotlib.pyplot as plt

#Set x-axis labels and data
objects = ('Facebook', 'Instagram', 'Twitter', 'Snapchat','YouTube')
y_pos = np.arange(len(objects))
performance = [35,15,1,25,40]

#Create bar plot with labels,a title, and color
plt.bar(y_pos, performance, align='center', alpha=0.5,color='yellow')
plt.xticks(y_pos, objects)
plt.xticks(rotation=45)
plt.ylabel('Daily Social Media Use (minutes)')
plt.xlabel('Applications')
plt.title('Daily Social Media Use Per Application')

#Display the bar plot
plt.show()
```

## Side-by-side bar plot

We are going to make a side-by-side bar plot that portrays the “Number of Sales in July and August per Week”. This code will define the data for the various bars and create a legend to show which color bar corresponds to which set of data. It will also create x- and y- labels for the plot as well as a title! 

```python
import numpy as np
import matplotlib.pyplot as plt

#First set of data
july_sales = (500, 800, 700, 900, 750)
ind = np.arange(5)  # the x locations for the groups
width = 0.35       # the width of the bars
fig, ax = plt.subplots()

#Creating the first set of bars for the first set of data
rects1 = ax.bar(ind, july_sales, width, color='blue') 

#Second set of data-- creation of bars 
august_sales = (800, 900, 600, 800, 1000)
rects2 = ax.bar(ind + width, august_sales, width, color='orange')

# Add some text for labels, title and axes ticks
ax.set_ylabel('Number of Sales',fontsize=13)
ax.set_title('Number of Sales in July and August per Week',fontsize=13)
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('Week 1', 'Week 2', 'Week 3', 'Week 4', 'Week 5'),fontsize=10)

#Create a legend 
ax.legend((rects1[0], rects2[0]), ('July', 'August'))

#Attach a text label above each bar displaying its height
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., height,
                '%d' % int(height),
                ha='center', va='bottom')
autolabel(rects1)
autolabel(rects2)
plt.show()
```

## Stacked bar plot


We are going to make a stacked bar plot that portrays the “Number of Sales in July and August per Week”. This code will define the data sets to plot and create the bars that will be stacked on top of each other. It will also create x- and y- labels, a title, and tick marks. This code will also create labels for each bar as well as a legend, so the number of sales per week can be shown clearly. 

```python
import numpy as np
import matplotlib.pyplot as plt

#Data
july_sales = (500, 800, 700, 900, 750)
august_sales = (800, 900, 600, 800, 1000)

#Setting the width of the bars and tick positions
width = 0.5      # the width of the bars
ind = np.arange(1,6)  # the x locations for the groups
tick_pos = [i + (width/50) for i in ind]

#Creating the bars
p1 = plt.bar(ind, july_sales, width, color='blue',align='center')
p2 = plt.bar(ind,august_sales,width,bottom=july_sales,color='green',align='center')

#Add some text for labels, title, axes ticks, and legend
plt.ylabel('Number of Sales',fontsize=12)
plt.xlabel('Weeks',fontsize=12,labelpad=15)
plt.title('Number of Sales in July and August per Week',fontsize=13)
plt.xticks(tick_pos,('Week 1','Week 2','Week 3','Week 4','Week 5'), fontsize=10)
plt.legend((p1[0], p2[0]), ('July', 'August'),loc="best")

#Labels for each bar
for r1, r2 in zip(p1, p2):
    h1 = r1.get_height()
    h2 = r2.get_height()
    plt.text(r1.get_x() + r1.get_width() / 2., h1 / 2., "%d" % h1, ha="center", va="bottom", color="white", fontsize=10, fontweight="bold")
    plt.text(r2.get_x() + r2.get_width() / 2., h1 + h2 / 2., "%d" % h2, ha="center", va="bottom", color="white", fontsize=10, fontweight="bold")

#Display the bar plot
plt.show()
```

## Histogram

We are going to make a histogram that portrays various IQs and the probability of people having these IQs. This code will create a best-fit line for the histogram and add it to the plot as soon as the data has been created. A grid and labels will also be added to the plot. 

<font color ='red'>Note the highlighted code below, this code uses a ‘normed = 1’ option and fixes the axes in the histogram. You likely don’t want to do with with yours. Be sure to remove these parts for your use, then your y axis will be counts in each bin rather than some uninterpretable number.</font>

```python

#!/usr/bin/env python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

#Used to create a best-fit line
mu, sigma = 100, 15
x = mu + sigma*np.random.randn(10000)

#Creates a histogram of the data
#This histogram has 30 bins, but it can be changed.
n, bins, patches = plt.hist(x, 30, normed=1, facecolor='green', alpha=0.75,edgecolor='black')

#Adds a 'best fit' line
y = mlab.normpdf(bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=1)

#Adds x- & y-labels for the history and creates a grid
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=15$')
#Highlighted code below
plt.axis([40, 160, 0, 0.03]) #Gives ranges for x & y values
#End of highlighted code
plt.grid(True)

#Shows the histogram
plt.show()
```
