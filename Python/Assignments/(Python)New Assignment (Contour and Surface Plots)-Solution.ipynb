# New Assignment- Contour and Surface Plots

In this assignment, create a contour plot and surface plot for the 2 equations below. One should look like a bowl and the other should look like a pringle. *Hint:* The code should be the same to generate both functions, the only thing you need to change will be the function itself.

## Z = (X^2)/(3^2) + (Y^2)/(3^2)
Generate just the contour plot for this function.

```python
import matplotlib.pyplot as plt
import numpy as np

#Creating the functions
def f(x,y):
    return (x**2)/(3**2) + (y**2)/(3**2)

#Defining x and y ranges for data set
x = np.linspace(-50,50,50)
y = np.linspace(-50,50,40)

X,Y = np.meshgrid(x,y)

#Z-values for the functions
Z = f(X,Y)

plt.figure()

cp = plt.contour(X,Y,Z)

plt.clabel(cp, inline=True, fontsize=10)
plt.title('Contour Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

Now, generate the surface plot.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

#Creating the figure
fig = plt.figure()
ax = fig.gca(projection='3d') #creating axes

#Creating the functions
def f(x,y):
    return (x**2)/(3**2) + (y**2)/(3**2)

#Defining x and y ranges for data set
x = np.linspace(-5.0,5.0,50)
y = np.linspace(-5.0,5.0,40)

X,Y = np.meshgrid(x,y)

#Z-values for the functions
R = f(X,Y)

Z = np.sin(R)

# Plot the surface.
#The surface is made opaque by using antialiased=False.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,linewidth=0.5, rstride=1,cstride=1,alpha=0.5,antialiased=False) 

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)

#Setting the tick locators-- choosing tick locations
#Creates evenly spaced ticks from min to max 
ax.zaxis.set_major_locator(LinearLocator(10)) 
#sprintf format string
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f')) 

# Add a color bar which maps values to colors.
fig.colorbar(surf,shrink=0.6,aspect =10,pad=0.09)

#Add labels for x,y, and z axes
ax.set_xlabel('X axis',fontsize=10)
ax.set_ylabel('Y axis',fontsize=10)
ax.set_zlabel('Z axis',fontsize=10)
ax.set_title('Cool-Warm Surface Plot',fontsize=10)
plt.show()
```

## Z = (Y^2)/(3^2) - (x^2)/(3^2)
Generate just the contour plot for this function.

```python
import matplotlib.pyplot as plt
import numpy as np

#Creating the functions
def f(x,y):
    return (y**2)/(3**2) - (x**2)/(3**2)

#Defining x and y ranges for data set
x = np.linspace(-50,50,50)
y = np.linspace(-50,50,40)

X,Y = np.meshgrid(x,y)

#Z-values for the functions
Z = f(X,Y)

plt.figure()

cp = plt.contour(X,Y,Z)

plt.clabel(cp, inline=True, fontsize=10)
plt.title('Contour Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

Now, generate the surface plot.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

#Creating the figure
fig = plt.figure()
ax = fig.gca(projection='3d') #creating axes

#Creating the functions
def f(x,y):
    return (y**2)/(3**2) - (x**2)/(3**2)

#Defining x and y ranges for data set
x = np.linspace(-5.0,5.0,50)
y = np.linspace(-5.0,5.0,40)

X,Y = np.meshgrid(x,y)

#Z-values for the functions
R = f(X,Y)

Z = np.sin(R)

#Setting rotation angle of plot
ax.view_init(azim=165) 

# Plot the surface.
#The surface is made opaque by using antialiased=False.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,linewidth=0.5, rstride=1,cstride=1,alpha=0.5,antialiased=False) 

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)

#Setting the tick locators-- choosing tick locations
#Creates evenly spaced ticks from min to max 
ax.zaxis.set_major_locator(LinearLocator(10)) 
#sprintf format string
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f')) 

# Add a color bar which maps values to colors.
fig.colorbar(surf,shrink=0.6,aspect =10,pad=0.09)

#Add labels for x,y, and z axes
ax.set_xlabel('X axis',fontsize=10)
ax.set_ylabel('Y axis',fontsize=10)
ax.set_zlabel('Z axis',fontsize=10)
ax.set_title('Cool-Warm Surface Plot',fontsize=10)
plt.show()
```
