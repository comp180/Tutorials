## 2. Run the Following Code
This code will red in all the images and find the percentage of red, green, and blue within each picture. It prints out the color and each percentage.

````r
library(OpenImageR)

pictures <- c("red.jpg", "blue.jpg", "cyan.jpg", "green.jpg", 
              "magenta.jpg", "yellow.jpg", "white.jpg")

for(pic in pictures) {
    cat(gsub(".jpg", "", pic), "\n")
    img2 <- readImage(pic)
    picArray <- img2@.Data
    
    red.mean <- mean(picArray[,,1])
    green.mean <- mean(picArray[,,2])
    blue.mean <- mean(picArray[,,3])
    
    total = red.mean + blue.mean + green.mean

    r.percent <- red.mean / total
    cat("Red Percent: ", r.percent)
    g.percent <- green.mean / total
    cat("\nGreen Percent: ", g.percent)
    b.percent <- blue.mean / total
    cat("\nBlue Percent: ", b.percent, "\n\n")
}
````
