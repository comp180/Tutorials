## Chicago Diversity Pie Chart
Create a pie chart that depicts the diversity of [Chicago](https://en.wikipedia.org/wiki/Demographics_of_Chicago).
* White- 32%
* Black- 32%
* Hispanic- 28%
* Asian- 5%
* Other- 3%

```python
import matplotlib.pyplot as plt

labels = ['White','Black','Hispanic','Asian','Other']
sizes = [32,32,28,5,3]
colors = ['red','orange','purple','blue','grey']

plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.title('Diversity of Chicago', fontweight='bold')
plt.axis('equal')
plt.show()
```

## Golden State Warriors' Wins Bar Plot
Create a bar plot for the number of wins per season that the [Golden State Warriors](https://www.basketball-reference.com/teams/GSW/) had since 2010. Be sure to include a title, labels, and colors.
* 2017-   67
* 2016-   73
* 2015-   67 
* 2014-   51
* 2013-   47
* 2012-   23
* 2011-   36
* 2010-   26

```python
import numpy as np 
import matplotlib.pyplot as plt

years = ['2017','2016','2015','2014','2013','2012','2011','2010']
y_pos = np.arange(len(years))
number_wins = [67,73,67,51,47,23,36,26]

rect = plt.bar(y_pos,number_wins,align='center',alpha=0.5,color='purple')

plt.xticks(y_pos,years)
plt.ylabel('Number of Wins')
plt.xlabel('Years')
plt.title('Number of Wins per Season for the Golden State Warriors')

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., height,
                '%d' % int(height),
                ha='center', va='bottom')
autolabel(rect)

plt.show()
```

## Adolescents Sports Group Bar Chart

Create a group bar chart for the percentage of male and female adolecents who [play sports](http://www.espn.com/espn/story/_/id/9469252/hidden-demographics-youth-sports-espn-magazine). Remember to include an apporpriate title. 
* Basketball
    * Male- 40%
    * Female- 25%
* Baseball/Softball
    * Male- 24%
    * Female- 17%
* Soccer
    * Male- 20%
    * Female- 17%
* Track
    * Male- 17%
    * Female- 17%
* Swimming
    * Male- 9%
    * Female- 12%
* Tennis
    * Male- 8%
    * Female- 8%
* Cross Country
    * Male- 5%
    * Female- 4%

```python
import numpy as np 
import matplotlib.pyplot as plt

male_sports = (40,24,20,17,9,8,5)
ind = np.arange(len(male_sports))
width = 0.35

fig,ax = plt.subplots()

#Set of bars for children sales
rects1 = ax.bar(ind,male_sports,width,color='pink')

#Set of bars for adult sales
female_sports = (25,17,17,17,12,8,4)
rects2 = ax.bar(ind+width,female_sports,width,color='red')

ax.set_ylabel('Percentage of Adolescents who Play Sports')
ax.set_xlabel('Sports')
ax.set_title('Percentage of Adolscents who Play Various Sports')
ax.set_xticks(ind+width/2)
ax.set_xticklabels(('Basketball','Baseball/Softball','Soccer','Track','Swimming','Tennis','Cross Country'))

ax.legend((rects1[0],rects2[0]),('Male','Female'))

#Attach a text label above each bar displaying its height
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., height,
                '%d' % int(height),
                ha='center', va='bottom')
autolabel(rects1)
autolabel(rects2)

plt.show()
```

```python
#!/usr/bin/env python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from scipy.stats import norm

ages = [83, 83, 82, 82, 82, 80, 80, 80, 79, 77, 76, 76, 76, 75, 74, 74, 73, 73, 73, 72, 72, 72, 71, 70, 70, 69, 69, 69, 69, 69, 68, 67, 67, 67, 66, 66, 66, 66, 66, 65, 65, 65, 65, 65, 64, 64, 64, 64, 64, 63, 63, 62, 62, 62, 61, 61, 61, 61, 61, 61, 61, 60, 60, 60, 59, 59, 59, 58, 58, 58, 57, 56, 56, 56, 56, 55, 55, 55, 54, 54, 53, 53, 53, 52, 52, 51, 49, 48, 48, 47, 46, 45, 45, 45, 45, 44, 44, 43, 42, 39]
num_bins = 15

mu, std = norm.fit(ages)

n, bins, patches = plt.hist(ages, num_bins, normed=True,facecolor='orange', alpha=0.75, edgecolor='black')

#Adding a best-fit line
xmin, xmax = plt.xlim()
x = np.linspace(xmin,xmax,100)
p = norm.pdf(x,mu,std)
plt.plot(x,p,linewidth=2)

plt.title('Distribution of Ages in Congress')
plt.xlabel('Ages in Congress',fontsize=14)

plt.show()
```
