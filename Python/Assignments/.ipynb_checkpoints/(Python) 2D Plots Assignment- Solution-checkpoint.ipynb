### 1. Linear Plots

X: 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014

Y1: 10000, 9500, 11100, 12000, 13300, 14000, 8450, 9010, 15788, 16000, 15500, 16890

Y2: 4500, 3980, 6002, 5789, 8000, 8350, 7540, 9011, 9450, 10100, 9200, 12300

#### A

Make a linear plot of X and Y1. Be sure to use markers. Use labels X and Y1 and give an appropriate title.

```python
import matplotlib.pyplot as plt

x = [2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014]
y1 = [10000, 9500, 11100, 12000, 13300, 14000, 8450, 9010, 15788, 16000, 15500, 16890]

plt.plot(x,y1,marker='o',color='g')

plt.title("Y1 vs. X")
plt.ylabel('Y1')
plt.xlabel('X')

plt.show()
```

#### B

Make a linear plot with X and Y1 and on the same axes plot X and Y2 using a different color line and markers. Add a legend.

```python
import matplotlib.pyplot as plt

x = [2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014]
y1 = [10000, 9500, 11100, 12000, 13300, 14000, 8450, 9010, 15788, 16000, 15500, 16890]
y2 = [4500, 3980, 6002, 5789, 8000, 8350, 7540, 9011, 9450, 10100, 9200, 12300]

plt.plot(x,y1,marker='o',color='g')
plt.plot(x,y2,marker='^',color='b')

plt.title("Y vs. X")
plt.ylabel('Y')
plt.xlabel('X')
plt.legend(['Y1','Y2'],loc='best')

plt.show()
```

#### A

To see how well a students’ GPAs predicts their motivation, let’s examine a scatter plot. Create a scatter plot of the GPAs on the x-axis and students’ motivation levels on the y- axis. Make sure markers are of sufficient size. Choose markers that can overlap (e.g. empty instead of filled markers).

```python
import matplotlib.pyplot as plt

#GPAs on the x-axis
gpa = [2.0, 2.0, 2.0, 2.0, 2.3, 2.6, 2.6, 2.8, 3.0, 3.1, 3.1, 3.2, 3.6, 3.8, 3.8, 3.9, 4.0]
#Motivation levels on the y-axis
motivation_levels = [50, 48, 100, 12, 34, 30, 78, 65, 87, 84, 75, 72, 83, 90, 90, 97, 98]

plt.scatter(gpa,motivation_levels,facecolors='none',edgecolors='red')

plt.xlabel('Student GPA')
plt.ylabel('Student Motivation')
plt.title('Student Motivation vs. Student GPA')

plt.show()
```

### 3. Subplot

#### A

Use the subplot feature of your language to put the following functions on the same figure:

Y1 = sin(2x)

Y2 = cos(2x)

Y3 = tan(2x)

```python
import matplotlib.pyplot as plt
import numpy as np

#X-values will go from -pi to pi
x = np.linspace(-2*np.pi, 2*np.pi)

#First plot sin(2x)
y1 = np.sin(2*x)
#Seond plot cos(2x)
y2 = np.cos(2*x)
#Third plot tan(2x)
y3 = np.tan(2*x)

#Creating first subplot for sin(2x)
plt.subplot(2,2,1)
plt.plot(x, y1)
plt.xlabel('Angle [rad]')
plt.ylabel('sin(2x)')

#Creating second subplot for cos(2x)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.xlabel('Angle [rad]')
plt.ylabel('cos(2x)')

#Creating third subplot for tan(2x)
plt.subplot(2,1,2)
plt.plot(x,y3)
#Creating axis so the full function of tan(2x) can be viewed with the asymptotes
plt.axis([-2*np.pi,2*np.pi,-2,2])
plt.xlabel('Angle [rad]')
plt.ylabel('tan(2x)')

plt.tight_layout()

plt.show()
```
