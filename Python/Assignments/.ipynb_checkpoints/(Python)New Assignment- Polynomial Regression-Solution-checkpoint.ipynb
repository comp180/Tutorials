# New Assignment- Polynomial Regression

We will be investigating the use of Polynomial Regression on the function x^3. We want to add some randomness to x^3, so we will be using the function below. Go ahead and run this code:

```python
import math
from numpy import *
import numpy as np
import random
import pandas as pd
from matplotlib.pyplot import *

def f(x):
    y = x**3 + 0.5 * np.random.normal(x,1,1)
    return y
```

Now we need to run this function. Make a list of 10 random test points between -2 and 2. Store this list as `x` and the pass `x` into `f` and store the results in `y`.

```python
x = []
for i in range(10):
    x.append(random.randint(-2,2))

y = []
for element in x:
    y.append(f(element))
```

Let's see the results. Make a data frame with x and y, and print out the data frame.

```python
x_df = pd.DataFrame(x)
y_df = pd.DataFrame(y)

print('Data Frame for x')
print(x_df)

print()
print()

print('Data Frame for y')
print(y_df)
```

Which model seems to fit or predict the data the best? Let's find out! First we need to create a test data set. Again, using the `randint()` function, store a list of 100 random values between -2 and 2, pass that list into `f()`, create a data frame named `testset`, and display `testset`.

```python
x_test = []
for i in range(100):
    x_test.append(random.randint(-2,2))
    
y_test = []
for element in x_test:
    y_test.append(f(element))
    
testset = pd.DataFrame(y_test)
print(testset)
```

Finally, find the root mean square error for all the polynomial models to find which one is the best. 

*Before you run the code, think which polynomial model will have the best results.*

```python
xtest_array = np.array(x_test)
ytest_array = np.array(y_test)

#Function for calculating the root mean square error
#The flatten() method is used to put the data types for the predictions
#and the targets in the same dimension so their values can be properly subtracted
def rmse(predictions, targets):
    return np.sqrt(np.mean((predictions.flatten()-targets.flatten())**2))

ytest_predicted = polyval(fit1, xtest_array)
print('Root mean square error for Degree 1')
rms1 = rmse(ytest_predicted,ytest_array)
print(rms1)

ytest_predicted = polyval(fit2, xtest_array)
print('Root mean square error for Degree 2')
rms2 = rmse(ytest_predicted,ytest_array)
print(rms2)

ytest_predicted = polyval(fit3, xtest_array)
print('Root mean square error for Degree 3')
rms3 = rmse(ytest_predicted,ytest_array)
print(rms3)

ytest_predicted = polyval(fit5, xtest_array)
print('Root mean square error for Degree 5')
rms5 = rmse(ytest_predicted,ytest_array)
print(rms5)

ytest_predicted = polyval(fit9, xtest_array)
print('Root mean square error for Degree 9')
rms9 = rmse(ytest_predicted,ytest_array)
print(rms9)
```
