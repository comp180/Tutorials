### 1. Pie Chart

Make a pie chart that shows the distribution of animals in a zoo that contains 30 penguins, 40 polar bears, 20 lions, and 10 otters. 

```python
import matplotlib.pyplot as plt

labels = ['Penguins','Polar Bears','Lions','Otters']
sizes = [30,40,20,10]
colors = ['red','orange','purple','blue']

plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.title('Distribution of Animals at the Zoo', fontweight='bold')
plt.axis('equal')
plt.show()
```

### 2. Bar Plot

Make a bar plot that shows the number of people, out of a total of 40, who like various movie types if 10 people like comedy; 8 people like action;  15 people like romance; 2 people like drama, and 5 people like sci-fi.

<font color="red">Be sure to include vertical or slanted labels for each bar on the X axis. </font>

```python
import numpy as np 
import matplotlib.pyplot as plt

type_movies = ['Comedy','Action','Romance','Drama','Sci-Fi']
y_pos = np.arange(len(type_movies))
people = [10,8,15,2,5]

plt.bar(y_pos,people,align='center',alpha=0.5,color='purple')

plt.xticks(y_pos,type_movies)
plt.ylabel('Number of People')
plt.xlabel('Movie Types')
plt.title('Number of People who Like Various Movie Types')

plt.show()
```

### 3. Side-by-side Bar Plot

Make a side-by-side bar plot that displays the number of children visitors and the number of adult visitors at a water park in the months of April, May, June and July.

<font color="red">Be sure to include vertical or slanted labels for the side-by-side bars on the X axis. Also, be sure to include a legend and a label on the Y axis. </font>


<b>April<br></b>
Children: 780<br>
Adults: 315<br>
<b>May<br></b>
Children: 1050<br>
Adults: 400<br>
<b>June<br></b>
Children: 3056<br>
Adults: 1000<br>
<b>July<br></b>
Children: 5025<br>
Adults: 1500

```python
import numpy as np 
import matplotlib.pyplot as plt

children_sales = (780,1050,3056,5025)
ind = np.arange(len(children_sales))
width = 0.35

fig,ax = plt.subplots()

#Set of bars for children sales
rects1 = ax.bar(ind,children_sales,width,color='pink')

#Set of bars for adult sales
adult_sales = (315,400,1000,1500)
rects2 = ax.bar(ind+width,adult_sales,width,color='red')

ax.set_ylabel('Number of Sales')
ax.set_xlabel('Months')
ax.set_title('Number of Children and Adult Sales in April, May, June and July')
ax.set_xticks(ind+width/2)
ax.set_xticklabels(('April','May','June','July'))

ax.legend((rects1[0],rects2[0]),('Child','Adult'))

#Attach a text label above each bar displaying its height
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., height,
                '%d' % int(height),
                ha='center', va='bottom')
autolabel(rects1)
autolabel(rects2)

plt.show()
```

### 4. Stacked Bar Plot

Make a stacked bar plot that displays the number of dark chocolate bars and the number of milk chocolate bars sold at the candy store in the months of May, June and July. <font color="red">Be sure to include vertical or slanted labels for the side-by-side bars on the X axis. Also, be sure to include a legend, a label on the Y axis, and labels on each of the bars.</font>

<b>May</b><br>
Dark chocolate: 2050<br>
Milk chocolate: 1750<br>
<b>June</b><br>
Dark chocolate: 4750<br>
Milk chocolate: 4000<br>
<b>July</b><br>
Dark chocolate: 4980<br>
Milk chocolate: 4650<br>


```python
import numpy as np
import matplotlib.pyplot as plt

#Data for the bar plots
dark_chocolate = (2050,4750,4980)
milk_chocolate = (1750,4000,4650)

#Width of bars and tick positions
width = 0.30
ind = np.arange(1,4)
tick_pos = [i+(width/50) for i in ind]

p1 = plt.bar(ind,dark_chocolate,width,color='purple',align='center')
p2 = plt.bar(ind,milk_chocolate,width,bottom=dark_chocolate,color='orange',align='center')

plt.ylabel('Chocolate Sales')
plt.xlabel('Months')
plt.title('Number of Dark Chocolate and Milk Chocolate Bars sold in May, June and July')
plt.xticks(tick_pos,('May','June','July'))

plt.legend((p1[0],p2[0]),('Dark Chocolate Bars','Milk Chocolate Bars'),loc='best')

#Labels for each bar
for r1, r2 in zip(p1, p2):
    h1 = r1.get_height()
    h2 = r2.get_height()
    plt.text(r1.get_x() + r1.get_width() / 2., h1 / 2., "%d" % h1, ha="center", va="bottom", color="white", fontsize=10, fontweight="bold")
    plt.text(r2.get_x() + r2.get_width() / 2., h1 + h2 / 2., "%d" % h2, ha="center", va="bottom", color="white", fontsize=10, fontweight="bold")

plt.show()
```

```python
#!/usr/bin/env python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from scipy.stats import norm

scores = [89.5,67.0,54.0,97.0,87.0,93.5,73.0,84.0,49.0,77.5,63.5,98.0,57.0,84.5,49.5,45.0,62.0,91.5,87.0,73.5]
num_bins = 10

mu, std = norm.fit(scores)

n, bins, patches = plt.hist(scores, num_bins, normed=True,facecolor='orange', alpha=0.75, edgecolor='black')

#Adding a best-fit line
xmin, xmax = plt.xlim()
x = np.linspace(xmin,xmax,100)
p = norm.pdf(x,mu,std)
plt.plot(x,p,linewidth=2)

plt.title('Distribution of Calculus II Exam Scores')
plt.xlabel('Scores',fontsize=14)
plt.ylabel('Number of Students',fontsize=14)

plt.show()
```
