### 1. Contour Plot

#### A

Make a contour plot. Make sure to add labels in the plot or a legend for colors on the contours. You can choose either a filled contour plot or colored lines, your choice.

```python
import matplotlib.pyplot as plt
import numpy as np

#Creating the functions
def f(x,y):
    return np.sin(np.pi*x) * np.sin(np.pi*x)

def g(x,y):
    return np.cos(np.pi*x) * np.cos(np.pi*x)

#Defining x and y ranges for data set
x = np.linspace(-50,50,50)
y = np.linspace(-50,50,40)

X,Y = np.meshgrid(x,y)

#Z-values for the functions
Z1 = f(X,Y)
Z2 = g(X,Y)

#Difference between Z-values
Z = (Z2-Z1)

plt.figure()

cp = plt.contour(X,Y,Z)

labels = ['line1', 'line2']
for i in range(len(labels)):
    cp.collections[i].set_label(labels[i])

plt.legend(loc='upper left')

plt.clabel(cp, inline=True, fontsize=10)
plt.title('Contour Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### 2. Surface Plot (or mesh plot)

#### A

Using the same data set as before, create a surface plot. Also be sure to choose an appropriate color mapping to help in interpretation. If you can’t make a surface plot, a mesh plot (where the surface is not filled in) will suffice.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

#Creating the figure
fig = plt.figure()
ax = fig.gca(projection='3d') #creating axes

#Creating the functions
def f(x,y):
    return np.sin(np.pi*x) * np.sin(np.pi*x)

def g(x,y):
    return np.cos(np.pi*x) * np.cos(np.pi*x)

x = np.linspace(-1.0,1.0,50)
y = np.linspace(-1.0,1.0,40)

X,Y = np.meshgrid(x, y)

#Z-values for the functions
Z1 = f(X,Y)
Z2 = g(X,Y)

#Difference between Z-values
R = Z2-Z1

Z = np.sin(R) 

# Plot the surface.
#The surface is made opaque by using antialiased=False.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,linewidth=0.5, rstride=1,cstride=1,alpha=0.5,antialiased=False) 

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)

#Setting the tick locators-- choosing tick locations
#Creates evenly spaced ticks from min to max 
ax.zaxis.set_major_locator(LinearLocator(10)) 
#sprintf format string
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f')) 

# Add a color bar which maps values to colors.
fig.colorbar(surf,shrink=0.6,aspect =10,pad=0.09)

#Add labels for x,y, and z axes
ax.set_xlabel('X axis',fontsize=10)
ax.set_ylabel('Y axis',fontsize=10)
ax.set_zlabel('Z axis',fontsize=10)
ax.set_title('Cool-Warm Surface Plot',fontsize=10)
plt.show()
```
