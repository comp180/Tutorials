# 1. Pie Chart

Pie charts can be made by storing the data(numbers) in a matrix,  putting the cell labels into a cell array, and passing both the matrix and the cell array into the built-in pie() function. A title can be added to the pie chart using the title() function, passing in the desired title as a string parameter.
Documentation: https://octave.sourceforge.io/octave/function/pie.html

```octave
X = [10 35 5 30 20];  
labels = {'Ruby','Python','C','Java','JavaScript'}; 
pie(X,labels)
title('Top Programming Languages Used at Loyola') 
```

# 2. Bar Plot

A bar plot in Octave can be made with the similar logic as before, meaning storing the data in  a matrix and storing the labels in a cell array.  However, to generate the actual bar plot, the matrix will be passed into the bar() function and then the labels can be placed using the set() function.
Documentation: https://octave.sourceforge.io/octave/function/bar.html 

```octave
weights = [75 91 105]; # weights to be plotted
names ={'Sam','Darry','Jason'} # names to be plotted
H=bar(weights) # plot weights
set(gca,'XTickLabel',names) # place names as labels
xlabel('Weight Lifter') # label x axis
ylabel('Bench Press Weight (lbs)') # label y axis
```

# 3. Group Bar Plot / Stacked Bar Plot

Group bar plots are easily performed in Octave, and it follows the same procedure that is done for creating a regular bar plot. The only difference is how you store the data in the matrix that is going to be used to generate the plot. In this example, the bar plots are grouped by 4 different quizzes and the data consists of grades for 3 different students. So the data was stored by listing the 3 scores for each quiz and separating each set of 3 scores by a semicolon.  A legend was generated using the legend() function, passing in the corresponding labels for each bar.


```octave
# Grouped Bar Plot
quizScores = [70 83 99; 87 93 91; 98 86 97; 95 92 93];  #quiz scores
names ={'#1','#2','#3','#4'}
bar(quizScores)
set(gca,'XTickLabel',names)
xlabel('Quiz') # label x axis
ylabel('Grade(points)') # label y axis
legend('Albert','George','Andrew')
```

A stacked bar plot can be made by adding a parameter to the bar() function which is ‘stacked’.

```octave
# Stacked Bar Plot
quizScores = [70 83 99; 87 93 91; 98 86 97; 95 92 93];  #quiz scores
names ={'#1','#2','#3','#4'}
bar(quizScores,'stacked') #line that turns plot to stacked bar plot
set(gca,'XTickLabel',names)
xlabel('Quiz') # label x axis
ylabel('Grade(points)') # label y axis
legend('Albert','George','Andrew')
```

# 4. Histogram

A histogram can be produced by  calling the hist() function, passing in the data as the first parameter, and the number of bins as the second parameter. In this example, the randn() function was called to produce 10000 random numbers.


```octave
hist (randn (10000, 1), 30);  #plot 10000 random numbers 
```
