## Solutions

## 1. Contour Plot 
A contour plot can be made by calling the contour() function, supplying a meshgrid and functions as parameters. 

```octave
x = linspace(-2*pi,2*pi);
y = linspace(0,4*pi);
[X,Y] = meshgrid(x,y);
Z = sin(X)+cos(Y);

figure
contour(X,Y,Z)
colormap hsv 
```

The numerical values corresponding to the different layers of the contour plot can be shown by sending a fourth and fifth parameter to the contour() function, which are the strings ‘ShowText’ and ‘on’ respectively. This is exemplified in the code below.

```octave
x = -2:0.2:2;
y = -2:0.2:3;
[X,Y] = meshgrid(x,y);
Z = X.*exp(-X.^2-Y.^2);
figure
contour(X,Y,Z,'ShowText','on')
colormap summer
```

## 2. Surface Plot
A surface plot can be generated by passing a meshgrid and other functions through the surf() function. The view() function allows for manipulation of the plot viewing angle.

```octave
[X,Y] = meshgrid(1:0.5:10,1:20);
Z = sin(X) + cos(Y);
surf(X,Y,Z)
colormap autumn
colorbar
view(0,45)   #change angle via view() function
view(0,90)
view(0,180)
view(0,360)
view(90,180)
view(45,90)
```
