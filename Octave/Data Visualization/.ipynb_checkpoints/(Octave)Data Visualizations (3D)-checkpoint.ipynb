# Data Visualization: 3D plots - contour maps and surface plots

## Demo instructions
Create the following charts. All plots should have titles, x-axis labels, y-axis labels and legends sufficiently sized for readability

1. Contour plot
    * With labels or a legend for colors on the contours
    * At least two different color mappings: e.g. hot/cold or black/white
    * Filled/Unfilled between the lines
2. Surface plots (& optional mesh plots)
    * Different angles/views
    * Color mappings

For help, reference: [Three-Dimensional Plots](https://octave.org/doc/v4.2.2/Three_002dDimensional-Plots.html)
## Solutions
## 1. Contour Plot 
A contour plot can be made by calling the [contour()](https://octave.sourceforge.io/octave/function/contour.html) function, supplying a [meshgrid](https://octave.sourceforge.io/octave/function/meshgrid.html) and functions as parameters. To get a filled contour map, use the [contourf()](https://octave.sourceforge.io/octave/function/contourf.html) function. There are more [colormap options](https://octave.sourceforge.io/octave/function/colormap.html) that you can explore as well.

```octave
x = linspace(-2*pi,2*pi);
y = linspace(0,4*pi);
[X,Y] = meshgrid(x,y);
Z = sin(X)+cos(Y);

figure
contourf(X,Y,Z)
title("sin(x) + cos(y)")
xlabel("X")
ylabel("Y")
colormap hsv 
```

The numerical values corresponding to the different layers of the contour plot can be shown by sending a fourth and fifth parameter to the contour() function, which are the strings ‘ShowText’ and ‘on’ respectively. This is exemplified in the code below.

```octave
x = -2:0.2:2;
y = -2:0.2:3;
[X,Y] = meshgrid(x,y);
Z = X.*exp(-X.^2-Y.^2);
figure
contour(X,Y,Z,'ShowText','on')
title("x * e^(-(x^2) - (y^2)")
xlabel("X")
ylabel("Y")
colormap summer
```

## 2. Surface Plot
A surface plot can be generated by passing a meshgrid and other functions through the [surf()](https://octave.sourceforge.io/octave/function/surf.html) function. The [view()](https://octave.sourceforge.io/octave/function/view.html) function allows for manipulation of the plot viewing angle.

```octave
[X,Y] = meshgrid(1:0.5:10,1:20);
Z = sin(X) + cos(Y);
surf(X,Y,Z)
colormap autumn
colorbar
title("sin(x) + cos(x)")
xlabel("X")
ylabel("Y")
view(0,45)   #change angle via view() function
view(0,90)
view(0,180)
view(0,360)
view(90,180)
view(45,90)
```
