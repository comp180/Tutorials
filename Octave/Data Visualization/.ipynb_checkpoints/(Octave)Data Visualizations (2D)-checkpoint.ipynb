# Solutions

## 1. Multiple Line Plot
A multiple line plot can be produced in Octave by using the plotyy() function; the desired line equations can be passed in as parameters and they will then appear on the plot.

Documentation: https://www.gnu.org/software/octave/doc/v4.2.0/Two_002dDimensional-Plots.html 

```octave
x = linspace (-1, 1, 201); # line spacing
ax = plotyy (x,sin(pi*x), x,10*cos(pi*x)); #plot
colormap ("default");
xlabel ("X"); #label x
ylabel (ax(1), "Sin(pi*x)"); # label y
ylabel (ax(2), "10*cos(pi*x)"); # label y
title("Graph of Waves"); # label title
```

## 2. Scatter Plot
A scatter plot can be made using  the [scatter()](https://octave.sourceforge.io/octave/function/scatter.html) function. The x line spacing, y function, and other optional parameters must be supplied as parameters to this function. The first 2 parameters are x and y coordinates. The 3rd parameter is an array for the size of the markers (an empty array makes the markers the default 36 square points). The 4th and final required parameter is an array of colors for the markers.

```octave
x = linspace(0,3*pi,200);
y = cos(x) + rand(1,200);
c = linspace(1,10,length(x));
Fit = polyfit(x,y,1);
plot(polyval(Fit,x))
scatter(x,y,[],c) #scatter function
title("200 points of cos(x) plus a Random Number Between 0 and 1")
xlabel("X")
ylabel("cos(x) + rand(1)")
```

## 3. Sub Plot
A sub plot can be created through the combined usage of the [fplot()](https://octave.sourceforge.io/octave/function/fplot.html) and [subplot()](https://octave.org/doc/v4.2.0/Multiple-Plots-on-One-Page.html) functions.
* The fplot() plots a function within a specified range.  
    * The parameters that can be passed into fplot() are the function and the range the function will be plotted over.
* The subplot() function divides the  plot area into subplots with a specified index.  
    * The parameters for subplot() are the rows, the columns, and the index.

```octave
subplot (2, 1, 1)
fplot (@sin, [-10, 10]);
title("sin(x)")
xlabel("X")
ylabel("sin(x)")
subplot (2, 1, 2)
fplot (@cos, [-10, 10]);
title("cos(x)")
xlabel("X")
ylabel("cos(x)")
```
